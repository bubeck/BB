```python

def get_all_edges(points):
    all_edges = []
    p = points[:-1]

    y_min, y_max, x_val, x_max, slope = 0, 0, 0, 0, 0
    #sklon 1 / m

    for i in range(0, len(points) - 1):
        if (points[i][1] < points[i + 1][1]):
            y_min = points[i][1]
            x_val = points[i][0]
            y_max = points[i + 1][1]
        else:
            y_min = points[i + 1][1]
            x_val = points[i + 1][0]
            y_max = points[i][1]
        
        x_max = max(points[i][0], points[i+1][0])

        if (y_max != y_min):
            slope = (points[i + 1][0] - points[i][0])/(points[i + 1][1] - points[i][1])
            all_edges.append([y_min, y_max, x_val, slope])

    return all_edges
```

```python
def set_global_tab(global_tab):    
    global_tab.sort(key = itemgetter(0,1,2))
    return global_tab
```

```python

def set_active_tab(global_tab):    
    #nastavte riadok skenovania
    scan_line = global_tab[0][0]

    active = []
    temp = []
    i = 0

    while(i < len(global_tab)):
        if(global_tab[i][0] == scan_line):
            temp = global_tab.pop(i)
            active.append(temp[1:4])
            i -= 1
        i +=1
    return global_tab, scan_line, active
```

```python
def filling_polygon(active, global_tab, scan_line):    
    lines = []

    while(active):
        mass_x = []
        # step 1
        # Drawing intersects
        for i in range(0, len(active) - 1):
            if(i % 2 == 0):
                lines.append([[int(active[i][1]) ,int(active[i+1][1])], 
                              [-scan_line, -scan_line]])

        # step 2    
        scan_line += 1

        # step 3    
        #asset cleanup if t_max <= scan line
        h = 0
        while(h < len(active)):
            if(active[h][0] == scan_line):
                active.pop(h)
                h -= 1
            h += 1


        # increment x val with slope
        for i in range(0, len(active)):
            active[i][1] += active[i][2]



        # step 5    
        #adding from global to asset
        temp = []
        i = 0

        while(i < len(global_tab)):
            if(global_tab[i][0] == scan_line):
                temp = global_tab.pop(i)
                active.append(temp[1:4])
                i -= 1
            i +=1


        # step 6        
        active.sort(key = itemgetter(1))

    return lines
```

```python
fig, ax = plt.subplots()

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
ci = 0

def draw_poly(points):
    global ci
    global ac
    
    all_edges = get_all_edges(points)
    if len(all_edges) == 0:
        print("no edges")
        return
    
    global_tab = set_global_tab(all_edges)
    global_tab, scan_line, active = set_active_tab(global_tab)

    #for i in range(len(global_tab)):
    #    print(i, global_tab[i])

    lines = filling_polygon(active, global_tab, scan_line)

    for i in range(0, len(points)-1):
        plt.plot((points[i][0],points[i+1][0]),
                 (-points[i][1], -points[i+1][1]), c = colors[ci % len(colors)])
    #if ci != 0:
    #    for l in lines:
    #        plt.plot(l[0], l[1], c = 'r')

    ci += 1
    ac = colors[ci % len(colors)]
    
points = []
points.append([0, 0])
points.append([0, 200])
points.append([200, 200])
points.append([200, 0])
points.append([0, 0])

draw_poly(points)
```
